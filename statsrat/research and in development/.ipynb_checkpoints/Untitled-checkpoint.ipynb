```python
import numpy as np
import statsrat as sr
from statsrat import rw
from statsrat.expr.predef.pvl_iti import spont_rec
from plotnine import ggplot, geom_point, geom_line, aes, stat_smooth, facet_wrap, annotate, scale_x_continuous
```

```python
basic = rw.model(name = 'basic_cfg2_intercept',
                  fbase = rw.fbase.cfg2_intercept,
                  fweight = rw.fweight.none,
                  lrate = rw.lrate.cnst,
                  aux = rw.aux.basic)

Kalman = rw.model(name = 'Kalman_cfg2_intercept',
                  fbase = rw.fbase.cfg2_intercept,
                  fweight = rw.fweight.none,
                  lrate = rw.lrate.from_aux_direct,
                  aux = rw.aux.Kalman)

trials = spont_rec.make_trials(schedule = 'experimental')
ds = basic.simulate(trials)
sr.learn_plot(ds, var = 'w') + scale_x_continuous(name = 'stage', breaks = [10, 120], labels = ['a', 'b'])
#+ scale_x_continuous(name = 'stage', breaks = [0, 150, 300, 550], labels = stage_labels)
```

```python
var = 'w'
ds_var = ds[var] 
dims = list(ds_var.dims)
dims.remove('t') # remove dimensions other than time ('t')
n_dims = len(dims)
df = ds_var.to_dataframe()
df = df.reset_index()

n_stage = len(np.unique(ds.stage_name))
stage_start = []
stage_labels = []
for s in range(n_stage):
    t = ds.t.loc[ds.stage == s].values
    start_point = t.min()
    stage_start += [start_point]
    print(ds.stage_name.loc[{'t': start_point}].values)
    stage_labels += [ds.stage_name.loc[{'t': start_point}].values]
print(stage_start)
print(stage_labels)
```
